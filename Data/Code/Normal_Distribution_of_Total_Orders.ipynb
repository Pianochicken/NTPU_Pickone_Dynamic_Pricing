```
import numpy as np
import pandas as pd

### data_space_sort 資料表按照space_id排序
data_space_sort = pd.read_csv("Real_Order.csv", encoding = 'utf-8')
data_space_sort
```

```
###偵測資料數目
number_of_data = len(data_space_sort['space_id'])
print("資料筆數: ", number_of_data)
```

#宣告space_sort資料表的資料結構

```
space_id = np.array(data_space_sort['space_id'])
creat_data_space = np.array(data_space_sort['created_at'])
actual_date_space = np.array(data_space_sort['date'])
actual_time_space = np.array(data_space_sort['start'])
###將creat日期與時間合併
actual_data_space = np.array(actual_date_space + ' ' + actual_time_space)
###宣告陣列儲存每個space的lead_time
lead_time_space = np.arange(space_id[number_of_data-1],dtype = float)
```

#計算各筆訂單的平均數，變異數，標準差

```
###mean
import datetime
import math
a = 0   #紀錄每筆order
space_number = space_id[number_of_data-1]-3   #只算到space_id(184)
counter = np.arange(space_number,dtype = int)
lead_time_total = 0
counter_total = 0

for i in range(1,space_number+1):
  ###初始化相關陣列
  while (space_id[a] == i): 
    date1 = actual_data_space[a]
    date2 = creat_data_space[a]
    start = datetime.datetime.strptime(date2, '%d/%m/%Y %H:%M:%S')
    end = datetime.datetime.strptime(date1, '%d/%m/%Y %H:%M')
    time = end - start
    ###將秒轉為小時  
    lead_time_total += time.total_seconds()/3600   
    counter_total += 1
    a += 1
    
    if(a == number_of_data):break
      
  else:
    a += 0
    
  if(a == number_of_data):break

mean = lead_time_total/counter_total   #實際只有147個space
print('平均數為: ',mean,' hours')
```

#(space) 平均數、變異數、標準差

```
import datetime
import math
###Variance
a = 0
total = 0
variance = 0
variance_total = 0
counter_total = 0
for i in range(1,space_number+1):   
  while (space_id[a] == i): 
    date1 = actual_data_space[a]
    date2 = creat_data_space[a]
    start = datetime.datetime.strptime(date2, '%d/%m/%Y %H:%M:%S')
    end = datetime.datetime.strptime(date1, '%d/%m/%Y %H:%M')
    time = end - start
    total = time.total_seconds()/3600
    variance_total += pow(total - mean,2)
    counter_total += 1
    a += 1  
    if(a == number_of_data):break
  else:
    a += 0   
    
  if(a == number_of_data):break
variance = variance_total/counter_total
print('變異數為: ',variance,' hours')

###Standard
standard = pow(variance,0.5)
print('標準差為: ',standard,' hours')


```

＃常態分配

```
import math
standard_n = np.arange(counter_total)
a = b = distribution_a = distribution_b = distribution_c = distribution_d = distribution_e = distribution_f = distance = 0
for i in range(1,space_number+1):   
  while (space_id[a] == i): 
    date1 = actual_data_space[a]
    date2 = creat_data_space[a]
    start = datetime.datetime.strptime(date2, '%d/%m/%Y %H:%M:%S')
    end = datetime.datetime.strptime(date1, '%d/%m/%Y %H:%M')
    time = end - start
    total = time.total_seconds()/3600
    distance = total-mean
    if(distance > 0):
      standard_n[b] = math.ceil((total - mean)/standard)
      b += 1
      a += 1
    elif(distance < 0):
      standard_n[b] = math.floor((total - mean)/standard)
      b += 1
      a += 1 
    if(a == number_of_data):break
  else:
    a += 0   
    
  if(a == number_of_data):break
for j in range(0,len(standard_n)):
  if(standard_n[j] <= -3):
    distribution_a += 1
  elif(standard_n[j] == -2):
    distribution_b += 1
  elif(standard_n[j] == -1):
    distribution_c += 1
  elif(standard_n[j] == 1):
    distribution_d += 1
  elif(standard_n[j] == 2):
    distribution_e += 1
  elif(standard_n[j] >= 3):
    distribution_f += 1
print('分配如下:分別為(-3以下, -2, -1, 1, 2, 3以上)個標準差內')
print((round((distribution_a/len(standard_n))*100, 6)),',',
      (round((distribution_b/len(standard_n))*100, 6)),',',
      (round((distribution_c/len(standard_n))*100, 6)),',',
      (round((distribution_d/len(standard_n))*100, 6)),',',
      (round((distribution_e/len(standard_n))*100, 6)),',',
      (round((distribution_f/len(standard_n))*100, 6)),)

```
