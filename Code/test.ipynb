```
import numpy as np
import pandas as pd
data = pd.read_csv("20191209_Combined_Real_Fake_order.csv 的副本",encoding = 'big5')
x = data.loc(axis = 1)[['locate_id', 'size', 'price', 'price_per_size','start_time_group','End_time_group','mrt', 'weekday', 'Is_holiday', 'status']]
print(len(x))
x
```

```
x1 = x[x['locate_id'] == 20]
x2 = x[x['locate_id'] == 9]
x1['status']-=1
y1 = x1['status']
x1 = x1.drop('status', axis = 1)
#x1 = x1.drop('locate_id', axis = 1)
#x1 = x1.drop('mrt', axis = 1)0
#x1

for i in range(1, 33):
  x3 = x[x['locate_id'] == i]
  print("Locate_id = ", i , 'how many:',len(x3))
```

```
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data_le=x1.copy()
data_le['locate_id'] = labelencoder.fit_transform(data_le['locate_id'])
data_le['start_time_group'] = labelencoder.fit_transform(data_le['start_time_group'])
data_le['End_time_group'] = labelencoder.fit_transform(data_le['End_time_group'])
data_le['mrt'] = labelencoder.fit_transform(data_le['mrt'])
data_le['weekday'] = labelencoder.fit_transform(data_le['weekday'])
data_le['Is_holiday'] = labelencoder.fit_transform(data_le['Is_holiday'])
x1 = data_le
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(x1, y1, 
                                                    test_size=0.2, 
                                                    random_state=0)

model1 = LogisticRegression().fit(X_train, y_train)
# print('Accuracy of Logistic regression classifier on training set: {:.2f}'
#      .format(model.score(X_train, y_train)))
# print('Accuracy of Logistic regression classifier on test set: {:.2f}'
#      .format(model.score(X_test, y_test)))
print(model1.score(X_train,y_train))
print(model1.score(X_test,y_test))
```
