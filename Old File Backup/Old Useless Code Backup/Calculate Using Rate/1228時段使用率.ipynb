```
import numpy as np
import pandas as pd

data_used = pd.read_csv("2019_12_21_time_online_maybe_3.0.csv", encoding = 'big5')
data_used
```

```
data_used_space_id = np.array(data_used['space_id'])
data_group1 = np.array(data_used['use_time_1'])
data_group2 = np.array(data_used['use_time_2'])
data_group3 = np.array(data_used['use_time_3'])
data_group4 = np.array(data_used['use_time_4'])
```

```
group1 = np.arange(184,dtype = float)
group2 = np.arange(184,dtype = float)
group3 = np.arange(184,dtype = float)
group4 = np.arange(184,dtype = float)
```

#計算各時段使用時數

```
a = 0
b = 0
g1 = 0
g2 = 0
g3 = 0
g4 = 0

for i in range(1,185):
  while (data_used_space_id[a] == i):   
    g1 += data_group1[a]
    g2 += data_group2[a]
    g3 += data_group3[a]
    g4 += data_group4[a]
    group1[b] = g1
    group2[b] = g2
    group3[b] = g3
    group4[b] = g4
   
    a += 1
    if(a == 129475):break
      
  else:
    b += 1
    g1 = 0  
    g2 = 0
    g3 = 0
    g4 = 0 
    group1[b] = g1
    group2[b] = g2
    group3[b] = g3
    group4[b] = g4
    
  if(a == 129475):break

for i in range(0,184):
  print('space:',i+1,' 時段1的使用時數:',group1[i],' 時段2的使用時數:',group2[i],' 時段3的使用時數:',group3[i],' 時段4的使用時數:',group4[i])

```

#計算各時段總時數

```
data_total = pd.read_csv("1226_group_total_time.csv", encoding = 'big5')
data_total
```

```
data_total_space_id = np.array(data_total['space_id'])
data_all_group1 = np.array(data_total['totaltime1'])
data_all_group2 = np.array(data_total['totaltime2'])
data_all_group3 = np.array(data_total['totaltime3'])
data_all_group4 = np.array(data_total['totaltime4'])
```

```
data_total_group1 = np.arange(184,dtype = float)
data_total_group2 = np.arange(184,dtype = float)
data_total_group3 = np.arange(184,dtype = float)
data_total_group4 = np.arange(184,dtype = float)
```

```
a = 0
b = 0
g1 = 0
g2 = 0
g3 = 0
g4 = 0

for i in range(1,185):
  while (data_total_space_id[a] == i):   
    g1 += data_all_group1[a]
    g2 += data_all_group2[a]
    g3 += data_all_group3[a]
    g4 += data_all_group4[a]
    data_total_group1[b] = g1
    data_total_group2[b] = g2
    data_total_group3[b] = g3
    data_total_group4[b] = g4
   
    a += 1
    if(a == 180):break
      
  else:
    if (data_total_space_id[a] == i+1):
      g1 = 0
      g2 = 0
      g3 = 0
      g4 = 0
      b += 1
      if(b == 184):break

    else:
      b += 1
      g1 = 0  
      g2 = 0
      g3 = 0
      g4 = 0 
      data_total_group1[b] = g1
      data_total_group2[b] = g2
      data_total_group3[b] = g3
      data_total_group4[b] = g4
      if(b == 184):break

for i in range(0,184):
  print('space:',i+1,' 時段1的總時數:',data_total_group1[i],' 時段2的總時數:',data_total_group2[i],' 時段3的總時數:',data_total_group3[i],' 時段4的總時數:',data_total_group4[i])

```

#計算各時段使用率

```
group1_Usage_rate = np.arange(184,dtype = float)
group2_Usage_rate = np.arange(184,dtype = float)
group3_Usage_rate = np.arange(184,dtype = float)
group4_Usage_rate = np.arange(184,dtype = float)
```

```
for i in range (0,184):
  group1_Usage_rate[i] = group1[i]/data_total_group1[i]
  group2_Usage_rate[i] = group2[i]/data_total_group2[i]
  group3_Usage_rate[i] = group3[i]/data_total_group3[i]
  group4_Usage_rate[i] = group4[i]/data_total_group4[i]
for j in range (0,184):
  print('space',j+1,'時段1的使用率:',group1_Usage_rate[j],'時段2的使用率:',group2_Usage_rate[j],'時段3的使用率:',group3_Usage_rate[j],'時段4的使用率:',group4_Usage_rate[j])
  ### nan值代表沒有那個space
  ### 0代表使用率為0
  
  
```
